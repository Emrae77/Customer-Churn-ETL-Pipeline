```python
# read the data in the unity caatlog into a dataframe
# for accounts

df_accounts = spark.read.format("csv")\
  .option("header", True)\
  .option("inferSchema", True)\
  .load("/Volumes/databrick_tutorial_1/bronze/raw_files/accounts/accounts.csv")
```

```python
df_accounts.printSchema()
```

```python
df_accounts.display(10)
```

```python
# for customers
df_customers = (
    spark.read.format("csv")
        .option("header", True)
        .option("inferSchema", True)
        .load("/Volumes/databrick_tutorial_1/bronze/raw_files/customers/customers.csv")
)

```

```python
df_customers.printSchema()
```

```python
df_customers.display(10)
```

```python
df_transactions = (
    spark.read.format("csv")
        .option("header", True)
        .option("inferSchema", True)
        .load("/Volumes/databrick_tutorial_1/bronze/raw_files/transactions/*.csv")
)
```

```python
df_transactions.printSchema()
df_transactions.display()
```

```python
# for interactions
df_interactions = (spark.read.format("csv")\
        .option("header", True)\
        .option("inferSchema", True)\
        .load("/Volumes/databrick_tutorial_1/bronze/raw_files/interactions/interactions.csv"))
df_interactions.printSchema()
df_interactions.display()
```

```python
def save_to_delta(df, table_name):
    df.write.format("delta").mode("overwrite").option("overwriteSchema", True).save(f"/databrick_tutorial_1.bronze/{table_name}")

save_to_delta(df_accounts, "accounts")
save_to_delta(df_customers, "customers")
save_to_delta(df_transactions, "transactions")
save_to_delta(df_interactions, "interactions")
```
